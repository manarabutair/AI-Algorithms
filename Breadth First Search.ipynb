<h5> <u> Problem Description: </u></h5>

Implement a function called `BSF()` to explore a given graph. `BSF()` receives a graph, initial state, and goal state and returns a list containing the parent of each node and flag indicating if the goal state is reached. You can use the code and Pseudocode of general graph search in Lecuter 5 and update it to consider the following: 

1. There is a goal state. 
2. You should use the `queue` package instead of using the python list.
3. You should store the parent of each visited node using a suitable data structure to use in Exercise 4.




`````



import queue

def BSF(graph, node, goalState):
    visited=[]
    parents={}
    q=queue.Queue()
    flag=False
    visited.append(node)
    q.put(node)
    while not q.empty():
      m=q.get()
      if m==goalState:
        flag=True
        break
      for neighbour in graph[m]:
         if neighbour not in visited:
           parents[neighbour]=m
           visited.append(neighbour)
           q.put(neighbour)
    return flag, parents     
    
`````
